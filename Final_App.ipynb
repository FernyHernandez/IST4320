## Storing Data in a Database
This is a quick way to create a SQL database using SQLite, a very basic database management system. It does NOT have all the features such as postgres or MySQL, but it'll work for now. Your choices for SQLite are: 
- INTEGER (a whole number), 
- TEXT (character text), 
- REAL (numbers with decimals, like a float),
- BLOB (any binary large object, like a .jpg)

The process to connect to a different database management system is similar  ðŸ˜€.


````python
# This will create a local database called database.db
# The database will have a table called app_data with 
# two columns: name and age
import sqlite3
sql_connect = sqlite3.connect('database.db')
cur = sql_connect.cursor() # creates a cursor object 

cur.execute('''
CREATE TABLE IF NOT EXISTS app_data (
    name TEXT,
    age INTEGER
    )
''')
sql_connect.commit()
````

````python
import tkinter as tk
main = tk.Tk() #this makes a plain window

name_var = tk.StringVar() #declares a string variable
age_var = tk.IntVar() #declares an integer variable

label_age = tk.Label(main, text="What is your age?",
                 font=('helvetica',14,'bold'))
age_entry = tk.Entry(main,
    textvariable = age_var,
    font=('calibre',14,'normal'))                    

label_name = tk.Label(main, text="What is your name?",
                 font=('helvetica',14,'bold'))
name_entry = tk.Entry(main, 
    textvariable = name_var, 
    font=('calibre',14,'normal'))


def getNameAge():
    # Gets the name and age data
    name = name_var.get()
    age = age_var.get()

    # What do you want to do with the name and age data?
    # Save your data to your database using parameterized queries
    query = "INSERT INTO app_data (name, age) VALUES (?, ?)"
    cur.execute(query, (name, age)) # if you only have one variable make sure you leave as (name,)
    sql_connect.commit()

    # Clear the name and age
    name_var.set("")
    age_var.set("")
    main.destroy()
    
submit_button = tk.Button(main,
    text="Submit",
    font=('calibre',14,'normal'),
    command = getNameAge)

label_age.pack()
age_entry.pack()
label_name.pack()
name_entry.pack()
submit_button.pack()
main.mainloop()

````

````python
### Create a small app using tkinter! 
#a. Use at least one label customization
#b. Use at least one button customization
#c. any other widget or customization :)
import tkinter as tk
from tkinter import ttk, Toplevel

def open_temperature_converter():
    popup = Toplevel(main)
    popup.title("Temperature Converter")

    # Label for instruction
    lbl_instruction = tk.Label(popup, text="Enter the temperature and select the unit:")
    lbl_instruction.pack(pady=(10, 0))

    # Entry widget for input temperature
    entry_temp = tk.Entry(popup, width=20)
    entry_temp.pack(pady=10)

    # ComboBox for selecting the unit
    unit_choices = ["Celsius", "Fahrenheit", "Kelvin"]
    combo_units_from = ttk.Combobox(popup, values=unit_choices, state="readonly")
    combo_units_to = ttk.Combobox(popup, values=unit_choices, state="readonly")
    combo_units_from.set("Celsius")  # default value
    combo_units_to.set("Fahrenheit")  # default value
    combo_units_from.pack(pady=(10, 0))
    combo_units_to.pack(pady=10)

    # Button to perform conversion
    btn_convert = tk.Button(popup, text="Convert", command=lambda: convert_temperature(entry_temp.get(), combo_units_from.get(), combo_units_to.get(), lbl_result))
    btn_convert.pack(pady=(5, 10))

    # Label to display the result
    lbl_result = tk.Label(popup, text="")
    lbl_result.pack(pady=(0, 10))
    
def convert_temperature(value, from_unit, to_unit, result_label):
    try:
        temp = float(value)
        if from_unit == "Celsius" and to_unit == "Fahrenheit":
            result = (temp * 9/5) + 32
        elif from_unit == "Celsius" and to_unit == "Kelvin":
            result = temp + 273.15
        else: 
            result = temp
        
        if result is not None:
        # Add other conversions as needed
            result_label.config(text=f"Converted temperature: {result:.2f} {to_unit}")
        else:
            result_label.config(text="Conversion not supported between these units.")
    except ValueError:
        result_label.config(text="Invalid input! Please enter a valid number.")

def open_length_converter():
    popup = Toplevel(main)
    popup.title("Length Converter")

    # Label for instruction
    lbl_instruction = tk.Label(popup, text="Enter the length and select the unit:")
    lbl_instruction.pack(pady=(10, 0))

    # Entry widget for input length
    entry_length = tk.Entry(popup, width=20)
    entry_length.pack(pady=10)

    # ComboBox for selecting the units
    unit_choices = ["Meters", "Kilometers", "Feet", "Miles"]
    combo_units_from = ttk.Combobox(popup, values=unit_choices, state="readonly")
    combo_units_to = ttk.Combobox(popup, values=unit_choices, state="readonly")
    combo_units_from.set("Meters")  # default value
    combo_units_to.set("Kilometers")  # default value
    combo_units_from.pack(pady=(10, 0))
    combo_units_to.pack(pady=10)

    # Button to perform conversion
    btn_convert = tk.Button(popup, text="Convert", command=lambda: convert_length(entry_length.get(), combo_units_from.get(), combo_units_to.get(), lbl_result))
    btn_convert.pack(pady=(5, 10))

    # Label to display the result
    lbl_result = tk.Label(popup, text="")
    lbl_result.pack(pady=(0, 10))
    
def convert_length(value, from_unit, to_unit, result_label):
    try:
        length = float(value)
        # Conversion factors: From Meters to any unit
        conversions = {
            "Meters": 1,
            "Kilometers": 0.001,
            "Feet": 3.28084,
            "Miles": 0.000621371
        }
        # Convert the input length to meters first
        meters = length / conversions[from_unit]
        # Convert from meters to the target unit
        result = meters * conversions[to_unit]
        result_label.config(text=f"Converted length: {result:.3f} {to_unit}")
    except ValueError:
        result_label.config(text="Invalid input! Please enter a valid number.")

def open_weight_converter():
    popup = Toplevel(main)
    popup.title("Weight Converter")

    # Label for instruction
    lbl_instruction = tk.Label(popup, text="Enter the weight and select the unit:")
    lbl_instruction.pack(pady=(10, 0))

    # Entry widget for input weight
    entry_weight = tk.Entry(popup, width=20)
    entry_weight.pack(pady=10)

    # ComboBox for selecting the units
    unit_choices = ["Grams", "Kilograms", "Pounds", "Ounces"]
    combo_units_from = ttk.Combobox(popup, values=unit_choices, state="readonly")
    combo_units_to = ttk.Combobox(popup, values=unit_choices, state="readonly")
    combo_units_from.set("Grams")  # default value
    combo_units_to.set("Kilograms")  # default value
    combo_units_from.pack(pady=(10, 0))
    combo_units_to.pack(pady=10)

    # Button to perform conversion
    btn_convert = tk.Button(popup, text="Convert", command=lambda: convert_weight(entry_weight.get(), combo_units_from.get(), combo_units_to.get(), lbl_result))
    btn_convert.pack(pady=(5, 10))

    # Label to display the result
    lbl_result = tk.Label(popup, text="")
    lbl_result.pack(pady=(0, 10))

def convert_weight(value, from_unit, to_unit, result_label):
    try:
        weight = float(value)
        # Conversion factors from grams to any unit
        conversions = {
            "Grams": 1,
            "Kilograms": 1000,
            "Pounds": 453.592,
            "Ounces": 28.3495
        }
        if from_unit in conversions and to_unit in conversions:
        # Convert the input weight to grams first
            grams = weight * conversions[from_unit]
        # Convert from grams to the target unit
            result = grams / conversions[to_unit]
        else:
            result = None
        
        if result is not None:
            result_label.config(text=f"Converted weight: {result:.3f} {to_unit}")
        else:
            result_label.config(text="Conversion not supported between these units.")
            
    except ValueError:
        result_label.config(text="Invalid input! Please enter a valid number.")
main = tk.Tk()

main.title("Unit Converter")
btn_temperature = tk.Button(main, text="Temperature", command=open_temperature_converter)
btn_length = tk.Button(main, text="Length", command=open_length_converter)
btn_weight = tk.Button(main, text="Weight", command=open_weight_converter)

btn_temperature.pack(pady=10)
btn_length.pack(pady=10)
btn_weight.pack(pady=10)

main.mainloop()
````
